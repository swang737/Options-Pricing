```python
import numpy as np
from matplotlib.pyplot import subplots
```

```python
# 1a
T = 1 # total time
d = 2 ** 10
dt = T / d # time step
u_0 = 1
u_1 = u_0 + 2 * dt
arr_u = [u_0, u_1] # output u array

for i in range(2, int(d / T) + 1):
    tn = (i-1) * dt
    arr_u.append((arr_u[i-1] * (dt * (1 + tn ** 2) + 2 - dt ** 2) - arr_u[i-2]) / (1 + dt * (1 + tn ** 2))) # formula
```

```python
fig, ax = subplots(figsize = (8, 8))
ax.plot(arr_u)
ax.set_xlabel('t')
ax.set_ylabel('u(t)')
```

```python
# 1b
T = 1 # total time
d = 2 ** 10
dt = T / d # time step
u_0 = 1
u_1 = u_0 + 2 * dt
arr_u = [u_0, u_1] # output u array

for i in range(2, int(d / T) + 1):
    tn = (i-1) * dt
    arr_u.append((arr_u[i-1] * (2 - dt ** 2) + arr_u[i-2] * ((dt * (1 + tn ** 2))/(2) - 1)) / ((dt * (1 + tn ** 2))/(2) + 1))
```

```python
fig, ax = subplots(figsize = (8, 8))
ax.plot(arr_u)
ax.set_xlabel('t')
ax.set_ylabel('u(t)')
```

```python
def getArrU(N):
    T = 1 # total time
    d = 2 ** N
    dt = T / d # time step
    u_0 = 1
    u_1 = u_0 + 2 * dt
    arr_u = [u_0, u_1] # output u array

    for i in range(2, int(d / T) + 1):
        tn = (i-1) * dt
        arr_u.append((arr_u[i-1] * (2 - dt ** 2) + arr_u[i-2] * ((dt * (1 + tn ** 2))/(2) - 1)) / ((dt * (1 + tn ** 2))/(2) + 1))

    return arr_u
```

```python
getArrU(10)[-1], getArrU(2)[-1]
```

```python
def getError(N):
    return getArrU(N - 1)[-1] -getArrU(N)[-1]
```

```python
getError(10)
```

```python
errors = []
for i in range (1, 17):
    error = getError(i)
    errors.append(error)

errors
```

```python
fig, ax = subplots(figsize = (8, 8))
ax.plot(-np.log(errors))
```
